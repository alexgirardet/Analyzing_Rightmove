## Preprocessing

````python
import numpy as np
````

````python
def handle_ids(x):
  count = 0

  query_results = x[0]
  query = x[1]
  insert = x[2]

  if query_results is not None:
    return 0
  elif query is not None:
    return 1
  elif insert is not None:
    return 2

def return_ids(x):

  query_results = x[0]
  query = x[1]
  insert = x[2]

  if query_results is not None:
    return query_results
  elif query is not None:
    return query
  elif insert is not None:
    return insert

def return_bytes(x):

  query_results = str(x[0])
  query = str(x[1])
  insert = str(x[2])

  if query_results != '<NA>':
    return float(query_results)
  elif query != '<NA>':
    return float(query)
  elif insert != '<NA>':
    return float(insert)
  else:
    return np.nan
````

````python
project_results = results[results['project_id'] == 'ric-eu-dl-ops-dev-nprd']
````

````python
project_results['owner'] = project_results[['jobGetQueryResultsResponse_jobId', 'jobQueryResponse_jobId', 'jobInsertResponse_jobId']].apply(handle_ids , axis=1)
````

````python
project_results['id'] = project_results[['jobGetQueryResultsResponse_jobId', 'jobQueryResponse_jobId', 'jobInsertResponse_jobId']].apply(return_ids , axis=1)
````

````python
project_results['bytes'] = project_results[['jobGetQueryResultsResponse_totalBilledBytes', 'jobQueryResponse_totalBilledBytes', 'jobInsertResponse_totalBilledBytes']].apply(return_bytes , axis=1)
````
