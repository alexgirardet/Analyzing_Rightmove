# ID ANALYSIS

We know these are mutually exclusive

````python
project_results = results[results['project_id'] == 'ric-eu-dl-ops-dev-nprd']
````

````python
project_results_ids = project_results[['jobGetQueryResultsResponse_jobId', 'jobQueryResponse_jobId', 'jobInsertResponse_jobId', 'jobGetQueryResultsResponse_totalBilledBytes', 'jobQueryResponse_totalBilledBytes', 'jobInsertResponse_totalBilledBytes']]
````

````python
project_results_ids_cut = project_results_ids.drop_duplicates()
````

````python
def handle_ids(x):
  count = 0

  query_results = x[0]
  query = x[1]
  insert = x[2]

  if query_results is not None:
    return 'query_result'
  elif query is not None:
    return 'query'
  elif insert is not None:
    return 'insert'
````

````python
def return_ids(x):
  count = 0

  query_results = x[0]
  query = x[1]
  insert = x[2]

  if query_results is not None:
    return query_results
  elif query is not None:
    return query
  elif insert is not None:
    return insert
````

````python
def return_bytes(x):
  count = 0

  query_results = str(x[0])
  query = str(x[1])
  insert = str(x[2])

  if query_results != '<NA>':
    return query_results
  elif query != '<NA>':
    return query
  elif insert != '<NA>':
    return insert
  else:
    return np.nan
````

````python
project_results_ids['owner'] = project_results_ids[['jobGetQueryResultsResponse_jobId', 'jobQueryResponse_jobId', 'jobInsertResponse_jobId']].apply(handle_ids , axis=1)
````

````python
project_results_ids['id'] = project_results_ids[['jobGetQueryResultsResponse_jobId', 'jobQueryResponse_jobId', 'jobInsertResponse_jobId']].apply(return_ids , axis=1)
````

````python
project_results_ids['bytes'] = project_results_ids[['jobGetQueryResultsResponse_totalBilledBytes', 'jobQueryResponse_totalBilledBytes', 'jobInsertResponse_totalBilledBytes']].apply(return_bytes , axis=1)
````
